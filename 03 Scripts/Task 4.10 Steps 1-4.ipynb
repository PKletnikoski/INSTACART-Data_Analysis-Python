#  Step 3 
## Create regional segmentation of the data

```python
# Defining regions

Northeast = ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']
Midwest = ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']
South = ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']
West = ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']
```

```python
# Assign States in 'state' column to a region

df_merge.loc[df_merge['state'].isin(Northeast), 'region'] = 'Northeast'
df_merge.loc[df_merge['state'].isin(Midwest), 'region'] = 'Midwest'
df_merge.loc[df_merge['state'].isin(South), 'region'] = 'South'
df_merge.loc[df_merge['state'].isin(West), 'region'] = 'West'
```

```python
# Check Totals

df_merge['region'].value_counts(dropna = False)
```

```python
df_merge.head()
```

### Check for variations in spending habits by region

```python
crosstab = pd.crosstab(df_merge['region'], df_merge['spending_flag'], dropna = False)
```

#### Create an exclusion flag for low-activity customers (customers with less than 5 orders) and exclude them from the data

```python
# Create a flag to assign an activity label customers based on number of orders

df_merge.loc[df_merge['max_order'] < 5, 'activity_flag'] = 'low activity customer'
```
